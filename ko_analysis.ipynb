```python
ko_df.index = ko_df.index.astype(str).str.replace('.1', '', regex=False)
ko_df
```

```python
ko_df = ko_df.apply(pd.to_numeric, errors='coerce')
```

```python
seq_ko_value = {}

# loop through each row in ko_df
for seq_id, row in ko_df.iterrows():
    for millet in seq_ko_dict.keys():
        if millet in seq_id:
            # only keep KOs that are >1  AND also in seq_ko_dict
            valid_kos = [
                (ko_id,int(row[ko_id])) for ko_id in ko_df.columns
                if row[ko_id] > 1 and ko_id in seq_ko_dict[millet]
            ]
            seq_ko_value[seq_id] = valid_kos
            break
seq_ko_value
```

```python
ko_function_map = {}

ko_ids_set = set(ko_ids)
# converting file to dict
with open("ko_list.txt") as f:
    for line in f:
        parts = line.strip().split("\t")
        if len(parts) == 2:
            ko_id, name = parts
            if ko_id in ko_ids_set:
                ko_function_map[ko_id] = name
ko_function_map
```

```python
import re
#creating dict with seq n their kos n funcs obtained from ko_function_map dict

def clean_function_description(description):
    if ';' in description:
        description = description.split(';', 1)[1]
    return description

seq_to_functions = {}

for seq_id, ko_list in seq_ko_value.items():
    functions = []
    for ko in ko_list:
        func = ko_function_map.get(ko[0])
        if func:
            clean_func = clean_function_description(func)
            functions.append((ko,clean_func))
    seq_to_functions[seq_id] = functions

seq_to_functions
```

```python
def ko_df_create(gene_id, entries):
    ec_class_map = {
        '1': 'Oxidoreductases',
        '2': 'Transferases',
        '3': 'Hydrolases',
        '4': 'Lyases',
        '5': 'Isomerases',
        '6': 'Ligases',
        '7': 'Translocases'
    }

    rows = []
    for ko_info, description in entries:
        ko, abundance = ko_info

        # Extract EC numbers from the description
        ec_matches = re.findall(r'\[EC:([^\]]+)\]', description)
        ec_ids = ec_matches[0].split() if ec_matches else None

        # Clean function description
        function = re.sub(r'\s*\[EC:[^\]]+\]', '', description).strip()

        # Map ECs to EC classes
        if ec_ids:
            ec_classes = {ec_class_map.get(ec.split('.')[0], 'Unknown') for ec in ec_ids}
            ec_id_str = ', '.join(ec_ids)
            ec_class_str = ', '.join(ec_classes)
        else:
            ec_id_str = None
            ec_class_str = None

        rows.append({
            'gene_id': gene_id,
            'ko_id': ko,
            'ko_abundance': abundance,
            'ko_function': function,
            'ec_id': ec_id_str,
            'ec_class': ec_class_str
        })

    return pd.DataFrame(rows)
```

```python
ko77 = ko_df_create('millet_PP355677', seq_to_functions['millet_PP355677'])
ko78 = ko_df_create('millet_PP355678', seq_to_functions['millet_PP355678'])
ko79 = ko_df_create('millet_PP355679', seq_to_functions['millet_PP355679'])
ko80 = ko_df_create('millet_PP355680',seq_to_functions['millet_PP355680'])
```

```python
# Get KO to pathway mapping from KEGG API ---
import requests
ko_pathway_resp = requests.get('http://rest.kegg.jp/link/pathway/ko')
ko_pathway_resp.raise_for_status()
ko_pathway_lines = ko_pathway_resp.text.strip().split('\n')
```

```python
#dict of KO to map ids
ko_pathway_map = {}
for line in ko_pathway_lines:
    ko_full, pathway_full = line.split('\t')
    ko = ko_full.replace('ko:', '').upper()
    pathway = pathway_full.replace('path:', '')
    ko_pathway_map.setdefault(ko, []).append(pathway)
ko_pathway_map
```

```python
#Get pathway names
pathway_list_resp = requests.get('http://rest.kegg.jp/list/pathway')
pathway_list_resp.raise_for_status()
pathway_list_lines = pathway_list_resp.text.strip().split('\n')
```

```python
#dict of map ids to pwy names
pathway_name_map = {}
for line in pathway_list_lines:
    pid_full, pname = line.split('\t')
    pid = pid_full.replace('path:', '')
    pathway_name_map[pid] = pname
pathway_name_map
```

```python
#get map ids of each ko_id n add to df
def get_pathway_info(ko):
    ko = ko.upper()
    pathways = ko_pathway_map.get(ko, [])
    if not pathways:
        return ''
    # Filter out pathways with 'Unknown' names
    valid_pathways = [(pid, pathway_name_map.get(pid, 'Unknown')) for pid in pathways]
    valid_pathways = [(pid, name) for pid, name in valid_pathways if name != 'Unknown']
    if not valid_pathways:
        return ''
    # Join as "pathway_id:pathway_name"
    return '; '.join(f"{pid}:{name}" for pid, name in valid_pathways)
```

```python
# Add a new column to df
ko77['pathways'] = ko77['ko_id'].apply(get_pathway_info)
ko78['pathways'] = ko78['ko_id'].apply(get_pathway_info)
ko79['pathways'] = ko79['ko_id'].apply(get_pathway_info)
ko80['pathways'] = ko80['ko_id'].apply(get_pathway_info)
```

```python
def rename_and_split_pathways(df, old_col='pathways', new_map_ids_col='map_ids', new_names_col='pathway_names'):
    # Rename old pathways column to new_map_ids_col first
    df = df.rename(columns={old_col: new_map_ids_col})

    def split_map_ids_and_names(map_ids_str):
        if pd.isna(map_ids_str) or map_ids_str.strip() == '':
            return '', ''

        parts = [p.strip() for p in map_ids_str.split(';')]
        ids = []
        names = []

        for part in parts:
            if ':' in part:
                mid, name = part.split(':', 1)
                ids.append(mid.strip())
                names.append(name.strip())
            else:
                # If no colon, just keep whole part as id and empty name
                ids.append(part)

        return '; '.join(ids), '; '.join(names)

    # Apply the function and create two new columns
    df[[new_map_ids_col, new_names_col]] = df[new_map_ids_col].apply(
        lambda x: pd.Series(split_map_ids_and_names(x))
    )

    return df
```

```python
ko77 = rename_and_split_pathways(ko77)
ko78 = rename_and_split_pathways(ko78)
ko79 = rename_and_split_pathways(ko79)
ko80 = rename_and_split_pathways(ko80)
```

```python
import requests

def fetch_kegg_brite_hierarchy():
    url = "http://rest.kegg.jp/get/br:ko00001"
    response = requests.get(url)
    text = response.text

    brite_map = {}
    current_class = None
    current_subclass = None

    for line in text.split('\n'):
        if line.startswith('A'):
            current_class = line.split(' ', 1)[1].strip()
        elif line.startswith('B '):
            parts = [p for p in line.split(' ') if p]
            current_subclass = ' '.join(parts[2:]).strip()
        elif line.startswith('C '):
            parts = line[2:].strip().split()
            if parts:
                pid = parts[0]
                if not pid.startswith('map'):
                    pid = 'map' + pid
                if pid.startswith('map999'):
                    continue  # skip unclassified
                brite_map[pid] = {
                    'class': current_class,
                    'subclass': current_subclass,
                }

    return brite_map

brite_class_map = fetch_kegg_brite_hierarchy()
```

```python
def map_brite_fields(mapids_str, brite_map):
    """Map pathways to their class and subclass only, input is semicolon-separated map IDs"""
    if pd.isna(mapids_str) or mapids_str.strip() == '':
        return '', ''

    # Split map ids by semicolon and strip whitespace
    map_ids = [mid.strip() for mid in mapids_str.split(';')]

    classes = set()
    subclasses = set()
    for mid in map_ids:
        if mid in brite_map:
            classes.add(brite_map[mid]['class'])
            subclasses.add(brite_map[mid]['subclass'])

    # Join multiple entries with semicolon
    return (

        '; '.join(sorted(subclasses)),
        '; '.join(sorted(classes))

    )
```

```python
# Apply mapping function to add new columns
ko77[['brite_subclass', 'brite_class']] = ko77['map_ids'].apply(
    lambda x: pd.Series(map_brite_fields(x, brite_class_map))
)
ko78[['brite_subclass', 'brite_class']] = ko78['map_ids'].apply(
    lambda x: pd.Series(map_brite_fields(x, brite_class_map))
)
ko79[['brite_subclass', 'brite_class']] = ko79['map_ids'].apply(
    lambda x: pd.Series(map_brite_fields(x, brite_class_map))
)
ko80[['brite_subclass', 'brite_class']] = ko80['map_ids'].apply(
    lambda x: pd.Series(map_brite_fields(x, brite_class_map))
)
```

```python
import requests
import json

# Step 1: Download enzyme.dat file
url = "https://ftp.expasy.org/databases/enzyme/enzyme.dat"
filename = "enzyme.dat"

r = requests.get(url)
with open(filename, "wb") as f:
    f.write(r.content)
```

```python
# Step 2: Parse enzyme.dat to get EC → function mapping
ec_to_function = {}
with open(filename, "r") as f:
    ec_num = None
    func = None
    for line in f:
        if line.startswith("ID"):
            ec_num = line.split()[1].strip()
        elif line.startswith("DE"):
            func = line[5:].strip().rstrip(".")
        elif line.startswith("//"):
            if ec_num and func:
                ec_to_function[ec_num] = func
            ec_num = None
            func = None
ec_to_function
```

```python
import re

def resolve_all_funcs(ec, ec_to_function, seen=None):
    """
    Resolves all valid final functions for an EC number,
    handling multiple and recursive 'Transferred entry' chains.
    Returns a list of function strings.
    """
    if seen is None:
        seen = set()

    if ec in seen:
        return []
    seen.add(ec)

    func = ec_to_function.get(ec)
    if not func:
        return []

    if func.startswith("Deleted entry"):
        return []

    elif func.startswith("Transferred entry"):
        # Extract all ECs mentioned
        target_ecs = re.findall(r'\d+\.\d+\.\d+\.\d+', func)
        results = []
        for target_ec in target_ecs:
            results.extend(resolve_all_funcs(target_ec, ec_to_function, seen.copy()))
        return results

    else:
        return [func]

# Build cleaned mapping with all valid functions (as list)
cleaned_ec_to_function = {}

for ec in ec_to_function:
    funcs = resolve_all_funcs(ec, ec_to_function)
    if funcs:
        cleaned_ec_to_function[ec] = funcs
cleaned_ec_to_function
```

```python

def add_ec_function_column(df, ec_to_function_dict):
    # Create the 'ec_function' column
    df = df.drop(columns=['gene_id'], errors='ignore')
    df['ec_function'] = df['ec_id'].map(
        lambda ec: '; '.join(ec_to_function_dict.get(ec, [])) if pd.notnull(ec) else ''
    )

    # Reorder columns: place 'ec_function' right after 'ec_class' if it exists
    if 'ec_class' in df.columns:
        cols = list(df.columns)
        idx = cols.index('ec_class') + 1
        cols.insert(idx, cols.pop(cols.index('ec_function')))
        df = df[cols]

    return df
```

```python
ko77 = add_ec_function_column(ko77, cleaned_ec_to_function)
ko78 = add_ec_function_column(ko78, cleaned_ec_to_function)
ko79 = add_ec_function_column(ko79, cleaned_ec_to_function)
ko80 = add_ec_function_column(ko80, cleaned_ec_to_function)
```
