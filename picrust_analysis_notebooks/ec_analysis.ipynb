```python
import requests
import json

# Step 1: Download enzyme.dat file
url = "https://ftp.expasy.org/databases/enzyme/enzyme.dat"
filename = "enzyme.dat"

r = requests.get(url)
with open(filename, "wb") as f:
    f.write(r.content)
```

```python
# Step 2: Parse enzyme.dat to get EC â†’ function mapping
ec_to_function = {}
with open(filename, "r") as f:
    ec_num = None
    func = None
    for line in f:
        if line.startswith("ID"):
            ec_num = line.split()[1].strip()
        elif line.startswith("DE"):
            func = line[5:].strip().rstrip(".")
        elif line.startswith("//"):
            if ec_num and func:
                ec_to_function[ec_num] = func
            ec_num = None
            func = None
ec_to_function
```

```python
import re

def resolve_all_funcs(ec, ec_to_function, seen=None):
    """
    Resolves all valid final functions for an EC number,
    handling multiple and recursive 'Transferred entry' chains.
    Returns a list of function strings.
    """
    if seen is None:
        seen = set()

    if ec in seen:
        return []
    seen.add(ec)

    func = ec_to_function.get(ec)
    if not func:
        return []

    if func.startswith("Deleted entry"):
        return []

    elif func.startswith("Transferred entry"):
        # Extract all ECs mentioned
        target_ecs = re.findall(r'\d+\.\d+\.\d+\.\d+', func)
        results = []
        for target_ec in target_ecs:
            results.extend(resolve_all_funcs(target_ec, ec_to_function, seen.copy()))
        return results

    else:
        return [func]

# Build cleaned mapping with all valid functions (as list)
cleaned_ec_to_function = {}

for ec in ec_to_function:
    funcs = resolve_all_funcs(ec, ec_to_function)
    if funcs:
        cleaned_ec_to_function[ec] = funcs
cleaned_ec_to_function
```

```python
# Create new mapping
ec_seq_to_functions = {}

for seq_id, ec_list in ec_seq_dict.items():
    functions = []
    for ec in ec_list:
        func = cleaned_ec_to_function.get(ec[0][3:])
        if func:
            functions.append((ec,func))
    ec_seq_to_functions[seq_id] = functions

# Result
ec_seq_to_functions
```

```python
# EC class mapping
ec_class_map = {
    '1': 'Oxidoreductases',
    '2': 'Transferases',
    '3': 'Hydrolases',
    '4': 'Lyases',
    '5': 'Isomerases',
    '6': 'Ligases',
    '7': 'Translocases'
}

# Get all millet items
millet_items = list(ec_seq_to_functions.items())

# Dictionary to store the DataFrames
millet_dfs = {}

# Loop through and create each DataFrame
for i, (millet_id, entries) in enumerate(millet_items):
    rows = []
    for ec_num, func_list in entries:
        func_str = '; '.join(func_list)
        rows.append({
            'ec_number': ec_num[0],
            'ec_abundance': ec_num[1],
            'ec_function': func_str
        })

    df = pd.DataFrame(rows)
    df['ec_class'] = df['ec_number'].str.extract(r'EC:(\d)')
    df['ec_class_name'] = df['ec_class'].map(ec_class_map)

    # Store in dictionary with key like "ec77", "ec78", etc.
    df_name = f'ec{77 + i}'
    millet_dfs[df_name] = df
```

```python
def add_ec_mapping_info(ko_df, ec_df):
    # Make copies to avoid modifying original DataFrames
    ko_df = ko_df.copy()
    ec_df = ec_df.copy()

    # Normalize: remove 'EC:' prefix from ec_df['ec_number']
    ec_df['ec_number_clean'] = ec_df['ec_number'].str.replace('EC:', '', regex=False)

    # Ensure columns are strings for matching
    ko_df['ec_id'] = ko_df['ec_id'].astype(str)
    ec_df['ec_number_clean'] = ec_df['ec_number_clean'].astype(str)

    # Create lookup dictionary
    ec_abundance_map = ec_df.set_index('ec_number_clean')['ec_abundance'].to_dict()

    # Map ec_abundance from ec_id
    ko_df['ec_abundance'] = ko_df['ec_id'].map(ec_abundance_map)

    # Add 'mapped' column
    ko_df['mapped'] = ko_df['ec_abundance'].apply(lambda x: 'Yes' if pd.notnull(x) else 'No')

    return ko_df
```

```python

def add_kos_to_ec_table(ec_df, ko_to_ec_file='ko_to_ec.txt'):

    # Load KO-EC mapping from file
    ko_ec_df = pd.read_csv(ko_to_ec_file, sep='\t', header=None, names=['ec','ko'])

    # Clean prefixes from both KO and EC columns
    ko_ec_df['ko'] = ko_ec_df['ko'].str.replace('ko:', '', regex=False)
    ko_ec_df['ec'] = ko_ec_df['ec'].str.replace('ec:', '', regex=False)

    # Group all KO IDs by EC number
    ec_to_kos = ko_ec_df.groupby('ec')['ko'].apply(lambda x: ','.join(sorted(set(x)))).reset_index()
    ec_to_kos.columns = ['ec_number_clean', 'ko_ids']

    # Clean 'EC:' prefix in the ec_df for matching
    ec_df = ec_df.copy()
    ec_df['ec_number_clean'] = ec_df['ec_number'].str.replace('EC:', '', regex=False)

    # Merge the KO IDs into the EC dataframe
    merged_df = pd.merge(ec_df, ec_to_kos, on='ec_number_clean', how='left')

    # Drop the helper column
    merged_df.drop(columns=['ec_number_clean'], inplace=True)

    return merged_df
```

```python
def mark_ec_kos_mapped(ec_df, ko_df):
    # Convert ko_df['ko_id'] to a set for fast lookup
    ko_set = set(ko_df['ko_id'].dropna().astype(str))

    def check_kos_mapped(ko_ids):
        if pd.isna(ko_ids) or not ko_ids:
            return 'no'
        # Split and check if any KO is in ko_set
        for ko in ko_ids.split(','):
            if ko.strip() in ko_set:
                return 'yes'
        return 'no'

    # Apply function to each row
    ec_df = ec_df.copy()
    ec_df['mapped'] = ec_df['ko_ids'].apply(check_kos_mapped)

    return ec_df
```

```python
def collect_all_kos(*dfs):

    all_kos = set()  # Use set to avoid duplicates

    for df in dfs:
        # Drop missing values in 'ko_ids' column
        for ko_str in df['ko_ids'].dropna():
            # Split the string and strip spaces
            ko_list = [ko.strip() for ko in ko_str.split(',') if ko.strip()]
            all_kos.update(ko_list)

    return list(sorted(all_kos))  # Return sorted list for consistency
```

```python
# Assuming df1, df2, df3, df4 all have a 'ko_ids' column
ko_ids = collect_all_kos(ec_77, ec_78, ec_79, ec_80)
ko_ids
```

```python
ko_function_map = {}

ko_ids_set = set(ko_ids)
# converting file to dict
with open("ko_list.txt") as f:
    for line in f:
        parts = line.strip().split("\t")
        if len(parts) == 2:
            ko_id, name = parts
            if ko_id in ko_ids_set:
                ko_function_map[ko_id] = name
ko_function_map
```

```python

import re

def add_clean_ko_functions(ec_df, ko_function_dict):
    """
    Adds a 'ko_functions' column to ec_df with clean KO descriptions,
    removing short names and [EC:...] tags from the KO function strings.
    """

    def extract_clean_functions(ko_ids):
        if pd.isna(ko_ids):
            return ''

        clean_functions = []
        for ko in ko_ids.split(','):
            ko = ko.strip()
            if not ko:
                continue
            desc = ko_function_dict.get(ko)
            if desc:
                # Remove short form (before the first semicolon, if present)
                desc = desc.split(';', 1)[-1] if ';' in desc else desc
                # Remove EC numbers like [EC:...]
                desc = re.sub(r'\s*\[EC:[^\]]+\]', '', desc)
                desc = desc.strip()
                if desc:
                    clean_functions.append(desc)

        return ', '.join(clean_functions)

    # Copy to avoid modifying original
    ec_df = ec_df.copy()
    ec_df['ko_functions'] = ec_df['ko_ids'].apply(extract_clean_functions)
    return ec_df
```

```python
# Get KO to pathway mapping from KEGG API ---
import requests
ko_pathway_resp = requests.get('http://rest.kegg.jp/link/pathway/ko')
ko_pathway_resp.raise_for_status()
ko_pathway_lines = ko_pathway_resp.text.strip().split('\n')
```

```python
#dict of KO to map ids
ko_pathway_map = {}
for line in ko_pathway_lines:
    ko_full, pathway_full = line.split('\t')
    ko = ko_full.replace('ko:', '').upper()
    pathway = pathway_full.replace('path:', '')
    ko_pathway_map.setdefault(ko, []).append(pathway)
ko_pathway_map
```

```python
#Get pathway names
pathway_list_resp = requests.get('http://rest.kegg.jp/list/pathway')
pathway_list_resp.raise_for_status()
pathway_list_lines = pathway_list_resp.text.strip().split('\n')
```

```python
#dict of map ids to pwy names
pathway_name_map = {}
for line in pathway_list_lines:
    pid_full, pname = line.split('\t')
    pid = pid_full.replace('path:', '')
    pathway_name_map[pid] = pname
pathway_name_map
```

```python
def add_pathways_to_ec_df(ec_df, ko_to_pathway_map, pathway_name_map):
    def get_pathways_from_kos(ko_ids):
        if pd.isna(ko_ids):
            return '', ''

        path_ids = set()
        for ko in ko_ids.split(','):
            ko = ko.strip()
            if not ko:
                continue
            # Get pathways for KO
            paths = ko_to_pathway_map.get(ko)
            if paths:
                path_ids.update(paths)

        # Remove duplicates and sort
        path_ids = sorted(path_ids)
        path_names = [pathway_name_map.get(pid, '') for pid in path_ids]
        path_names = [name for name in path_names if name]  # Remove empty

        return ', '.join(path_ids), ', '.join(path_names)

    ec_df = ec_df.copy()
    ec_df[['pathway_ids', 'pathway_names']] = ec_df['ko_ids'].apply(
        lambda x: pd.Series(get_pathways_from_kos(x))
    )
    return ec_df
```

```python
ec_77 = add_pathways_to_ec_df(ec_77, ko_pathway_map, pathway_name_map)
ec_78 = add_pathways_to_ec_df(ec_78, ko_pathway_map, pathway_name_map)
ec_79 = add_pathways_to_ec_df(ec_79, ko_pathway_map, pathway_name_map)
ec_80 = add_pathways_to_ec_df(ec_80, ko_pathway_map, pathway_name_map)
```

```python
import requests

def fetch_kegg_brite_hierarchy():
    url = "http://rest.kegg.jp/get/br:ko00001"
    response = requests.get(url)
    text = response.text

    brite_map = {}
    current_class = None
    current_subclass = None

    for line in text.split('\n'):
        if line.startswith('A'):
            current_class = line.split(' ', 1)[1].strip()
        elif line.startswith('B '):
            parts = [p for p in line.split(' ') if p]
            current_subclass = ' '.join(parts[2:]).strip()
        elif line.startswith('C '):
            parts = line[2:].strip().split()
            if parts:
                pid = parts[0]
                if not pid.startswith('map'):
                    pid = 'map' + pid
                if pid.startswith('map999'):
                    continue  # skip unclassified
                brite_map[pid] = {
                    'class': current_class,
                    'subclass': current_subclass,
                }

    return brite_map

brite_class_map = fetch_kegg_brite_hierarchy()
```

```python
#map ids with its brite class subclass
brite_class_map
```

```python
def map_brite_fields(mapids_str, brite_map):
    """Map pathways to their class and subclass only, input is comma-separated map IDs"""
    if pd.isna(mapids_str) or mapids_str.strip() == '':
        return '', ''

    # Split map ids by semicolon and strip whitespace
    map_ids = [mid.strip() for mid in mapids_str.split(',')]

    classes = set()
    subclasses = set()
    for mid in map_ids:
        if mid in brite_map:
            classes.add(brite_map[mid]['class'])
            subclasses.add(brite_map[mid]['subclass'])

    # Join multiple entries with semicolon
    return (

        '; '.join(sorted(subclasses)),
        '; '.join(sorted(classes))

    )
```

```python
# Apply mapping function to add new columns
ec_77[['brite_subclass', 'brite_class']] = ec_77['pathway_ids'].apply(
    lambda x: pd.Series(map_brite_fields(x, brite_class_map))
)
ec_78[['brite_subclass', 'brite_class']] = ec_78['pathway_ids'].apply(
    lambda x: pd.Series(map_brite_fields(x, brite_class_map))
)
ec_79[['brite_subclass', 'brite_class']] = ec_79['pathway_ids'].apply(
    lambda x: pd.Series(map_brite_fields(x, brite_class_map))
)
ec_80[['brite_subclass', 'brite_class']] = ec_80['pathway_ids'].apply(
    lambda x: pd.Series(map_brite_fields(x, brite_class_map))
)
```
