```python
import pandas as pd
import requests
```

```python
abun = pd.read_csv("/content/path_abun_unstrat.tsv", sep="\t")
```

```python
pwys_abun = pd.DataFrame(abun)
```

```python
pwys_abun
```

```python
# consider pwys only which have abundance values greater than 0.1
cutoff = 0.1
seq_pwy_abun = {}
for seq in pwys_abun.columns:
    if seq != "pathway":
        pwy_list = pwys_abun[pwys_abun[seq] > cutoff]["pathway"].tolist()
        seq_pwy_abun[seq] = pwy_list
seq_pwy_abun
```

```python
#based on seq_pwy_abun dict obtained consider only these pwys from the report file n constuct a df
def filter_pathways(filename, pwy_dict, seq):

    allowed_pwys = set(pwy_dict.get(seq, []))
    cols = []

    with open(filename, "r") as f:
        for line in f:
            if line.startswith("path"):
                parts = line.strip().split()
                path_id = parts[parts.index("name") + 1]

                if path_id not in allowed_pwys:
                    continue

                fam0 = int(parts[parts.index("fam0") + 1])
                fam_found = int(parts[parts.index("fam-found") + 1])

                cols.append({
                    "Pathway": path_id,
                    "fam_total": fam0,
                    "fam_found": fam_found,
                    "completeness": round(fam_found / fam0, 2) if fam0 > 0 else 0
                })

    return pd.DataFrame(cols)
```

```python
# apply this func to each report file n sort them by abundance and
# then consider only pwys which have completeness>0.5
```

```python
pwys_77 = filter_pathways("/content/millet_PP355677_minpath_report.txt", seq_pwy_abun, "millet_PP355677")
pwys_77 = pwys_77.sort_values(by="completeness", ascending=False)
pwys_77=pwys_77[pwys_77['completeness']>0.5]
```

```python
pwys_78 = filter_pathways("/content/millet_PP355678_minpath_report.txt", seq_pwy_abun, "millet_PP355678")
pwys_78 = pwys_78.sort_values(by="completeness", ascending=False)
pwys_78=pwys_78[pwys_78['completeness']>0.5]
```

```python
pwys_79 = filter_pathways("/content/millet_PP355679_minpath_report.txt", seq_pwy_abun, "millet_PP355679")
pwys_79 = pwys_79.sort_values(by="completeness", ascending=False)
pwys_79=pwys_79[pwys_79['completeness']>0.5]
```

```python
pwys_80 = filter_pathways("/content/millet_PP355680_minpath_report.txt", seq_pwy_abun, "millet_PP355680")
pwys_80 = pwys_80.sort_values(by="completeness", ascending=False)
pwys_80=pwys_80[pwys_80['completeness']>0.5]
```

```python
# get pwy name from metacyc n add to df
```

```python
import gzip

url = "https://raw.githubusercontent.com/picrust/picrust2/master/picrust2/default_files/description_mapfiles/metacyc_pathways_info.txt.gz"
resp = requests.get(url)
resp.raise_for_status()

with open("metacyc_pathways_info.txt.gz", "wb") as f:
    f.write(resp.content)

pathway_map = {}

with gzip.open("metacyc_pathways_info.txt.gz", "rt") as f:
    for line in f:
        parts = line.strip().split("\t", 1)
        if len(parts) == 2:
            pid, desc = parts
            pathway_map[pid] = desc
```

```python
pathway_map
```

```python
pwys_77['pathway_name'] = pwys_77['Pathway'].map(pathway_map)
pwys_78['pathway_name'] = pwys_78['Pathway'].map(pathway_map)
pwys_79['pathway_name'] = pwys_79['Pathway'].map(pathway_map)
pwys_80['pathway_name'] = pwys_80['Pathway'].map(pathway_map)
```

```python
pwys_77.to_csv('pwy_77.csv', index=False)
pwys_78.to_csv('pwy_78.csv', index=False)
pwys_79.to_csv('pwy_79.csv', index=False)
pwys_80.to_csv('pwy_80.csv', index=False)
```
